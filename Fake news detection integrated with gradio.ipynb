```python
#separating data and label
x=df.drop(columns='label',axis=1)
y=df['label']
print(x)
print(y)
```

## Stemming Unit

```python
ps=PorterStemmer()
```

```python
def stemming(content):
  stemmed_content=re.sub('[^a-zA-z]',' ',content)
  stemmed_content=stemmed_content.lower()
  stemmed_content=stemmed_content.split()
  stemmed_content=[ps.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
  stemmed_content=' '.join(stemmed_content)
  return stemmed_content
```

```python
df['content']=df['content'].apply(stemming)
print(df['content'])
```

```python
#separating data and label after stemming
x=df['content'].values
y=df['label'].values
print("Content:",x)
print("Label:",y)
print("ROWS,COLUMNS:",x.shape[0],",",y.shape[0])
```

```python
print("ROWS,COLUMNS:",df.shape)
```

## Vectorizing Unit

```python
vectorizer=TfidfVectorizer()
vectorizer.fit(x)

x=vectorizer.transform(x)
print(x)
```

## Training and Fitting the model

1.spliting train and test **data**

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=3)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

2.Accuracy scores

```python
#getting accuracy scores of the model
#train data
x_train_prediction=model.predict(x_train)
train_accuracy=accuracy_score(x_train_prediction,y_train)
print("ACCURACY SCORE OF THE MODEL ON TRAINING DATA:",train_accuracy)
#test data
x_test_prediction=model.predict(x_test)
test_accuracy=accuracy_score(x_test_prediction,y_test)
print("\nACCURACY SCORE OF THE MODEL ON Test DATA:",test_accuracy)
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
import joblib

# Load the pre-trained model and vectorizer
model = joblib.load('trained_model.pkl')
vectorizer = joblib.load('tfidf_vectorizer.pkl')

# Function to preprocess and predict the label for a single line news article
def predict_fake_news_single_line(article):
    # Preprocess the article
    stemmed_content = preprocess_article(article)

    # Vectorize the preprocessed article
    vectorized_content = vectorizer.transform([stemmed_content])

    # Make prediction
    prediction = model.predict(vectorized_content)

    # Return the predicted label
    return prediction[0]

# Function to preprocess the article
def preprocess_article(article):
    ps = PorterStemmer()
    stemmed_content = re.sub('[^a-zA-Z]', ' ', article)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [ps.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content


```

```python
def predictor(title):
  predicted_label = predict_fake_news_single_line(title)
  if predicted_label == 0:
    return "REAL"
  else:
    return "FAKE"
```
